```python
import pandas as pd

target_file = "i-code.txt"
code_segment_length = 7 # Code always has 7 characters with 1 space after
with open(target_file) as f:
    df_dict = {'chapter_code': [], 'category_code': [], 'subcategory_code': [], 'title': []}
    for full_line in f:
        full_code = full_line[0:8]
        title = full_line[8:].rstrip()
        code = full_code.rstrip()
        
        # Ignore any code longer than 4 as the DB schema for code only holds up to 4-digits.
        if len(code) > 4:
            continue
        elif len(code) == 4:
            print(code)
            chapter_code = code[0]
            category_code = code[1:3]
            subcategory_code = code[3]

            df_dict['chapter_code'].append(chapter_code)
            df_dict['category_code'].append(category_code)
            df_dict['subcategory_code'].append(subcategory_code)
            df_dict['title'].append(title)
        else:
            chapter_code = code[0]
            category_code = code[1:3]
            subcategory_code = '-'

            df_dict['chapter_code'].append(chapter_code)
            df_dict['category_code'].append(category_code)
            df_dict['subcategory_code'].append(subcategory_code)
            df_dict['title'].append(title)

    df = pd.DataFrame.from_dict(df_dict)
    df.to_csv("i-icd-codes.csv", sep='|', index=False)
```
